# lmplot()

regression plot is used in machine learnign so we will not fully discuss what , how and where it is used but we will see how it is created and what are the values we can put inside it

in this pridction goes to y axis and pridction through goes on x axis

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)   # want to pridict tip
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',markers=['o','v'])
```
