** we use distribution plot to visualize the distribution of quantative data.
    lets discuss some plots that allow us to visualize the distribution of data set. These plots are-
    distplot/histplot
    joinplot
    pairplot
    rugplot**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=sns.load_dataset("tips")   # this comes from seaborn dataset it has many dataset (search from google)
df
```

```python
 df["size"].unique()
```

```python
sns.displot(df["total_bill"])
```

```python
sns.histplot(df["total_bill"],bins=6)     # hist and dist are same
```

```python
sns.histplot(df["total_bill"],bins=6,kde=True)   # kernel density estimation
```

```python
plt.subplot(1,2,1)
sns.histplot(df["total_bill"],bins=6,kde=True)
plt.subplot(1,2,2)
sns.histplot(df["tip"],bins=6,kde=True)
```

** join plot **

```python
sns.jointplot(x ='total_bill', y ='tip',data =df,kind= 'scatter',color='gold')
```

```python
sns.jointplot(x ='total_bill', y ='tip',data =df,kind= 'kde',color='gold')
```

```python
sns.jointplot(x ='total_bill', y ='tip',data =df,kind= 'hex',color='gold')
```

```python
sns.jointplot(x ='total_bill', y ='tip',data =df,kind= 'reg',color='gold')
```

** pairplot **

```python
sns.pairplot(df,hue='size',palette='rainbow')
```

** rugplot **

```python
sns.rugplot(df)
```
