```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

**Matplotlib Basics**

```python
x = np.linspace(0,5,11)
x
```

```python
y=x**2
y
```

```python
plt.plot(x,y)
plt.title("Title")
plt.xlabel("x axis")
plt.ylabel("y label")
```

**Subplots in Matplotlib**

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.subplot(2,2,2)
plt.plot(x**2,y**1)
plt.subplot(2,2,3)
plt.plot(y,x)
```

**Objected Oriented Way**

```python
fig = plt.figure()
axes1= fig.add_axes([0.1,0.1,1,1])
axes1.plot(x,y)
axes2=fig.add_axes([0.3,0.4,0.3,0.2])
axes2.plot(x,y)
```

**Types of Plots**

```python
plt.scatter(x,y)
```

```python
plt.hist(x)
```

```python
plt.boxplot(x)
```

**Savinng plots**

```python
fig = plt.figure()
axes= fig.add_axes([0.1,0.1,1,1])
axes.plot(x,y)
plt.savefig("basicplot.png")
```
